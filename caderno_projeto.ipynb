## Questão 13
Quais as variáveis que mais impactaram no valor da compra? Como você chegou a essa conclusão?

Criando a tabela para analisar as variáveis:

```python
pt_1 = df[['Purchase','Product_Category_1']]
pt_1 = pt_1.rename(columns={'Product_Category_1': 'categoria'})
pt_2 = df[['Purchase','Product_Category_2']]
pt_2 = pt_2.rename(columns={'Product_Category_2': 'categoria'})
pt_3 = df[['Purchase','Product_Category_3']]
pt_3 = pt_3.rename(columns={'Product_Category_3': 'categoria'})
pt_unido = pd.concat([pt_1, pt_2, pt_3])

pt_unido
```

Vamos analisar o fator de influência de cada grandeza através do seu fator de variação (medida padronizada de dispersão de uma distribuição de probabilidade ou de uma distribuição de frequências).

Extraindo o fator para cada uma das colunas:

```python
ept = pt_unido.groupby(by = ['categoria'])['Purchase'].mean()
print(ept.std()/ept.mean())
```

```python
ept
```

```python
print(a_1.std()/a_1.mean())
```

```python
print(b_1.std()/b_1.mean())
```
